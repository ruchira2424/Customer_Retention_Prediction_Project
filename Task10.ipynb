```python
import pandas as pd 
import numpy as np 
```

```python
train_df=pd.read_csv("churn-bigml-80.csv")
test_df=pd.read_csv("churn-bigml-20.csv")
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df=train_df.drop(columns=["State","Area code"])
test_df=test_df.drop(columns=["State","Area code"])
```

```python
from sklearn.preprocessing import LabelEncoder 

for col in ["International plan", "Voice mail plan", "Churn"]:
    le = LabelEncoder()
    train_df[col] = le.fit_transform(train_df[col])  
```

```python
fe = LabelEncoder()
test_df["International plan"]=fe.fit_transform(test_df["International plan"])
test_df["Voice mail plan"]=fe.fit_transform(test_df["Voice mail plan"])
test_df["Churn"]=fe.fit_transform(test_df["Churn"])
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(train_df.drop("Churn",axis=1))
x_test=scaler.fit_transform(test_df.drop("Churn",axis=1))

y_train=train_df["Churn"].values
y_test=test_df["Churn"].values
```

```python
x_train
```

```python
from tensorflow import keras 
from tensorflow.keras import layers 

y_train = y_train.astype(int)
y_test = y_test.astype(int)


model=keras.Sequential([
    layers.Dense(32,activation='relu',input_shape=(x_train.shape[1],)),
    layers.Dense(16,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history=model.fit(
    x_train,y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_loss,test_accuracy=model.evaluate(x_test,y_test)
```
